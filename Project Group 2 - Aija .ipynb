```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the data into a Pandas DataFrame

data_csv_path = "/gun_violence_data=pd.read_csv("archive.zip")
usa_gun_df=pd.read_csv(data_csv_path)
```

```python
#Total Number of Deaths from Mass Shootings in Each state.
mass_kill=usa_gun_df[usa_gun_df.n_killed >4].groupby("state").sum()
mass_kill_df =pd.DataFrame(data=mass_kill["n_killed"])
mass_chart=mass_kill_df.plot(kind="bar", title= "Total Deaths vs State",figsize=(12,12))
mass_chart.set_xlabel("State")
x_axis=np.arange(0,len("state"))
tick_locations=[]
for x in x_axis:
    tick_locations.append(x)
mass_chart.set_ylabel("Total Mass Shooting Deaths")
plt.show(20)




```

```python
#Total Number of Mass shootings in Each State. 
mass_kill=usa_gun_df[usa_gun_df.n_killed >4].groupby("state").count()
mass_kill_df =pd.DataFrame(data=mass_kill["n_killed"])
mass_kill_df
mass_chart=mass_kill_df.plot(kind="bar", title= "Number of Mass Shootings vs State",figsize=(12,12))
mass_chart.set_xlabel("State")
x_axis=np.arange(0,len("state"))
tick_locations=[]
for x in x_axis:
    tick_locations.append(x)
mass_chart.set_ylabel("Mass Shootings")
plt.show(20)
```

```python
#Total killings (mass and non mass) - top 6 states.
total_kill=usa_gun_df.groupby("state").sum().sort_values(by="n_killed", ascending=False)["n_killed"]
total_kill.head(6)
```

```python
#Total Mass Killing Injuries and Deaths by State
mass_injkill=usa_gun_df[usa_gun_df.n_killed >4].groupby("state").sum()
mass_injkill=mass_injkill["n_killed"]+mass_kill["n_injured"]


mass_injkillchart=mass_injkill.plot(kind="bar", title= "Number Deaths + Injuries vs State",figsize=(12,12))
mass_injkillchart.set_xlabel("State")
x_axis=np.arange(0,len("state"))
tick_locations=[]
for x in x_axis:
    tick_locations.append(x)
mass_injkillchart.set_ylabel("Deaths + Injuries")
plt.show(20)
```
